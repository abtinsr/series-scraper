```python
URL = "https://www.fifa.com/fifa-world-ranking/women?dateId=ranking_20210625"
page = requests.get(URL)

soup = bs(page.content, "lxml")
```

```python
cards = soup.find_all("tr", class_="fc-ranking-item-full_rankingTableFullRow__1nbp7 ")
```

```python
for card in cards:
    print(card['aria-label'], end="\n"*3)
```

```python
for card in cards:
    rank = card.find("h6", class_="ff-m-0")
    country = card.find("span", class_="d-none d-lg-block")
    
    print(rank.text)
    print(country.text, end="\n"*2)
```
