```python
df = episodes.query('series_name in ("Breaking Bad")')

sns.set(rc={'figure.figsize':(30,30)})
sns.set_style('white')
sns.set_context("poster")

# Plot each year's time series in its own facet
g = sns.relplot(
    data=df,
    x="episode", y="episode_rating", col="season", hue="season",
    kind="line", palette="crest", linewidth=4, zorder=5,
    col_wrap=2, height=4, aspect=2, legend=False,
)

# Iterate over each subplot to customize further
for season, ax in g.axes_dict.items():

    # Add the title as an annotation within the plot
    ax.text(.05, .1, season, transform=ax.transAxes, fontweight="bold")

    # Plot every year's time series in the background
    sns.lineplot(
        data=df, x="episode", y="episode_rating", units="season",
        estimator=None, color=".7", linewidth=1, ax=ax,
    )

# Reduce the frequency of the x axis ticks
ax.set_xticks(ax.get_xticks()[::2])

# Tweak the supporting aspects of the plot
g.set_titles("")
g.set_axis_labels("Episode", "Average Rating")
g.tight_layout()
```

```python
episodes['total_rating'] = episodes['episode_rating'] * episodes['episode_total_votes']
```

```python
seasons = (
    episodes
    .query('series_n_seasons != 1')
    .groupby(['series_name', 'season'])
    .aggregate({'episode_total_votes': 'sum', 'total_rating': 'sum'})
    .reset_index()
    .assign(average_rating=lambda x: round(x['total_rating'] / x['episode_total_votes'],1))
    .groupby(['series_name'])
    .apply(lambda x: x.assign(previous_average_rating=round(x['average_rating'].shift(1))))
    #.assign(previous_average_rating=lambda x: x['average_rating'].shift(1))
    .drop(columns=['total_rating', 'series_name'])
    .reset_index()
    .drop(columns=['level_1'])
    .assign(rating_change=lambda x: round(x['average_rating'] / x['previous_average_rating'],2)-1)
)
seasons['rating_change'] = seasons['rating_change'].fillna(0)
```

```python
sns.set(rc={'figure.figsize':(30,30)})
sns.set_style('white')
sns.set_context("poster")

# Plot each year's time series in its own facet
g = sns.relplot(
    data=seasons,
    x="season", y="average_rating", col="series_name", hue="series_name",
    kind="line", palette="crest", linewidth=4, zorder=5,
    col_wrap=2, height=5, aspect=2, legend=False,
)

# Iterate over each subplot to customize further
for series_name, ax in g.axes_dict.items():
    
    # Add the title as an annotation within the plot
    ax.text(.05, .1, series_name, transform=ax.transAxes, fontweight="bold")

    # Plot every year's time series in the background
    sns.lineplot(
        data=seasons, x="season", y="average_rating", units="series_name",
        estimator=None, color=".9", linewidth=2, ax=ax,
    )

# Reduce the frequency of the x axis ticks
ax.set_xticks(ax.get_xticks()[::1])

# Tweak the supporting aspects of the plot
g.set_titles("")
g.set_axis_labels("Season", "Average Rating")
g.tight_layout()
```
